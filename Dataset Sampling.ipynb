# Creating samples of the dataset
- We will create 6 samples
- Each sample will have all 66,646 features
- Each sample consists of equal number of datapoints belonging to each class.
- Every sample consists of 180 datapoints, with 20 datapoints for each class.
- Saved as pickle file.

```python
import os
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import numpy as np
```

Loading the dataset

```python
filename=open("cleaneddataset.pkl",'rb')
dataset=pickle.load(filename)
filename.close()
```

```python
dataset.head()
```

Creating Samples

```python
for i in tqdm(range(0,6)):
    # where key is the class name and value is the count of that class in our sample
    countdict={}
    # initialize it with every value being one
    for j in range(1,10):
        countdict[j]=0
    #list in which we will store 
    listofindexes=[]
    while True:
        randint=np.random.randint(0,10868)
        if countdict[dataset['Class'][randint]]==20:
            continue
        else:
            countdict[dataset['Class'][randint]]+=1
            listofindexes.append(randint)
        if len(listofindexes)==180:
            break
    
    pickle.dump(dataset.iloc[listofindexes],open("SampleNo."+str(i),'wb'))
```
